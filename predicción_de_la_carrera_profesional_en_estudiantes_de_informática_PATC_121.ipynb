

 Visualizamos la distribución de las variables categóricas con gráficos de barras, esto nos permite ver cuáles son las carreras más comunes, intereses, especialidades, etc.


```python
import pandas as pd
import matplotlib.pyplot as plt

categorical_cols = ['Gender', 'Major', 'Interested Domain', 'Future Career']


bar_color = '#72B9D2'


for col in categorical_cols:
    print(f"\nConteo de valores únicos para {col}:\n")
    print(df[col].value_counts())

    plt.figure(figsize=(12, 6))
    df[col].value_counts().sort_values(ascending=False).plot(
        kind='bar', color=bar_color, edgecolor='black'
    )
    plt.xlabel(col, fontsize=12)
    plt.ylabel("Frecuencia", fontsize=12)
    plt.title(f"Distribución de {col}", fontsize=14, weight='bold')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Este código recorre las columnas categóricas especificadas en 'categorical_cols' y para cada una de ellas:
# 1. Imprime el conteo de valores únicos, lo que nos ayuda a entender la distribución de las categorías.
# 2. Crea un gráfico de barras para visualizar la frecuencia de cada valor en la columna.
# El uso de gráficos de barras facilita la comparación visual de las frecuencias de las categorías,
# lo que es útil para detectar patrones, distribuciones desequilibradas o categorías poco representadas en el conjunto de datos.
# Este análisis es esencial en la exploración de datos antes de realizar cualquier modelado,
# ya que proporciona una visión clara sobre la variabilidad y los posibles sesgos en las variables categóricas.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


sns.set(style="whitegrid")
plt.rcParams["figure.dpi"] = 120


hist_color = "#EF476F"
violin_color = "#D62598"


numeric_cols = df.select_dtypes(include=['number']).columns

for col in numeric_cols:
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))


    sns.histplot(df[col], kde=True, ax=axes[0], color=hist_color)
    axes[0].set_title(f'Histograma de {col}', fontsize=12)
    axes[0].set_xlabel(col)
    axes[0].set_ylabel('Frecuencia')

    sns.violinplot(y=df[col], ax=axes[1], color=violin_color)
    axes[1].set_title(f'Gráfico de Violín de {col}', fontsize=12)
    axes[1].set_ylabel(col)

    plt.tight_layout()
    plt.show()

# Usamos histogramas y diagramas de violín para explorar la distribución de variables numéricas como la edad y el GPA
# Esto ayuda a identificar sesgos, outliers y patrones
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/kkarliw/cs_students-1-/refs/heads/main/cs_students%20(1).csv"
df = pd.read_csv(url)

plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 12

columns_to_drop = ['Unnamed: 32', 'id']
existing_columns_to_drop = [col for col in columns_to_drop if col in df.columns]

if existing_columns_to_drop:
    df = df.drop(columns=existing_columns_to_drop, axis=1)

numeric_cols = df.select_dtypes(include=['number']).columns

plt.figure(figsize=(12, 10))
sns.heatmap(df[numeric_cols].corr(),
            annot=True,
            fmt=".1f",
            linewidths=.5,
            annot_kws={"size": 8},
            cmap="Reds")
plt.title('Mapa de Calor de Correlación de Atributos Numéricos', fontsize=14)
plt.show()

# El mapa de calor generado es una excelente herramienta para visualizar las relaciones entre las características numéricas en el dataset.
# Las correlaciones altas (cerca de 1 o -1) indican una relación fuerte entre dos variables, mientras que las correlaciones cercanas a 0 indican que no existe una relación lineal significativa.
# Este análisis puede ayudar a identificar patrones importantes que podrían ser útiles en modelos predictivos o en la toma de decisiones.
```
